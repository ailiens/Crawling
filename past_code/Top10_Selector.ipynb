```python
import pandas as pd

# 데이터 읽기
file_path = '0521_Combined_NaverBlogs.xlsx'
df = pd.read_excel(file_path)

# '플랫폼'이 '네이버'인 데이터 필터링 및 '인플루언서 유무'가 '인플루언서'인 행 필터링
# filtered_df = df[(df['플랫폼'] == '네이버') & (df['인플루언서 유무'] == '인플루언서')]
# 
# # 상위 10개 행 추출
# top_10_naver_df = filtered_df.head(10)

# '검색 방법' 컬럼 추가
df['검색 방법'] = df['날짜'].apply(lambda x: '인플루언서' if x == 'No date found' else '블로그')

# 결과를 새로운 엑셀 파일로 저장
filtered_file_path = 'top10_Naver.xlsx'
df.to_excel(filtered_file_path, index=False)

print('저장완료')
```
