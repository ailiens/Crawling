```python
import pandas as pd

def load_data(file_path):
    """엑셀 파일 로드"""
    return pd.read_excel(file_path)

def get_titles_by_authors(data, authors):
    """주어진 작성자 리스트에 대한 제목 추출"""
    filtered_data = data[data['블로거'].isin(authors)]
    return filtered_data[['블로거', '제목', '검색어']]

def print_titles(titles_df):
    """제목 데이터 프레임 출력"""
    for author in titles_df['블로거'].unique():
        print(f"블로거: {author}")
        author_titles = titles_df[titles_df['블로거'] == author][['제목', '검색어']]
        for _, row in author_titles.iterrows():
            print(f"- {row['제목']} (검색어: {row['검색어']})")
        print("\n")

# 파일 경로 설정
file_path = '0426_Combined_NaverBlogs.xlsx'

# 데이터 로드
data = load_data(file_path)

# 분석할 작성자 리스트
authors = ['청비 대제',
           'Barista7',
           '앱테크로 커피한잔',
           '이든',
           '그런이의 슬기로운 커피생활',
           '조망키',
           '김비실씨 오늘은 뭐 했어요?',
           '제이디',
           '우박',
           '기억리을٩(◕ᗜ◕)و',
           '새벽리더',
           '곰곰이이야기',
           '인포머',
           '나마카',
           'Zeroback Game Issue',
           '궈니'
           ]

# 작성자별 제목 가져오기
titles_by_authors = get_titles_by_authors(data, authors)

# 제목 출력
print_titles(titles_by_authors)
```

```python
import pandas as pd
from collections import Counter

def analyze_data(df):
    # 작성자와 제목을 기준으로 그룹화하고, 각 그룹에서 유일한 검색어를 추출
    grouped = df.groupby(['블로거', '제목'])['검색어'].unique()

    # 같은 작성자와 제목을 가지면서 여러 검색어로 나온 결과를 나열
    multiple_queries = grouped[grouped.apply(len) > 1]

    # 모든 검색어를 하나의 리스트에 수집
    all_queries = []
    for queries in multiple_queries:
        all_queries.extend(queries)
    
    # 검색어 빈도수 계산
    query_frequencies = Counter(all_queries)

    # 빈도수를 기준으로 검색어를 정렬
    sorted_queries_by_frequency = sorted(query_frequencies.items(), key=lambda item: item[1], reverse=True)

    return multiple_queries, sorted_queries_by_frequency

# 데이터 로드 (샘플 데이터, 실제 사용 시 파일 로드로 대체)
data = pd.read_excel('0426_Combined_NaverBlogs.xlsx')
df = pd.DataFrame(data)

# 데이터 분석 실행
multiple_search_results, search_query_stats = analyze_data(df)

# 결과 출력
print("Multiple search queries per author and title:")
print(multiple_search_results)
print("\nSearch query frequencies sorted by frequency:")
for query, count in search_query_stats:
    print(f"{query}: {count}")
```

```python
import pandas as pd
from collections import Counter

# 데이터 로드
df = pd.read_excel('0426_Combined_NaverBlogs.xlsx')

# 검색어 빈도 계산
all_queries = []
for index, row in df.iterrows():
    if isinstance(row['검색어'], list):
        all_queries.extend(row['검색어'])
    elif isinstance(row['검색어'], str):
        # 문자열 검색어가 쉼표로 구분된 경우
        all_queries.extend(row['검색어'].split(', '))

query_frequencies = Counter(all_queries)

# 각 행별로 가장 빈도가 높은 검색어의 빈도를 계산
def get_max_frequency(row):
    if isinstance(row['검색어'], list):
        return max(query_frequencies[query] for query in row['검색어'])
    elif isinstance(row['검색어'], str):
        return max(query_frequencies[query] for query in row['검색어'].split(', '))

df['max_query_frequency'] = df.apply(get_max_frequency, axis=1)

# 결과 확인
print(df[['블로거', '제목', '검색어', 'max_query_frequency']].sort_values(by='max_query_frequency', ascending=False))
```
