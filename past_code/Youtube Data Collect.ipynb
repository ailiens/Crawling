```python
from selenium.common import TimeoutException
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Selenium WebDriver 설정 (Chrome 사용)
s = Service("C:\\chromedriver\\chromedriver.exe")
driver = webdriver.Chrome(service=s)

# YouTube에 접속
driver.get("https://www.youtube.com/@GameTalkTalk")
time.sleep(2)  # 페이지 로드를 위한 대기 시간

# Shorts 탭 클릭
shorts_button = driver.find_element(By.XPATH, '//*[@id="tabsContent"]/yt-tab-group-shape/div[1]/yt-tab-shape[3]/div[1]')
shorts_button.click()
time.sleep(2)

# 스크롤 다운하여 추가 콘텐츠 로드
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(2)  # 스크롤 후 페이지 로드 대기
print('스크롤 완료')

# 제목 수집
titles_elements = driver.find_elements(By.XPATH, '//*[@id="video-title"]')
titles = [title.text.strip() if title.text.strip() else "직접기입" for title in titles_elements]
print("Collected Titles:", titles)

# 데이터 저장을 위한 리스트 초기화
all_video_data = []

# 각 비디오 별로 추가 정보 수집
videos_xpath = '//*[@id="dismissible"]'
videos = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, videos_xpath)))

for i, video in enumerate(videos):
    video.click()  # 비디오 클릭
    time.sleep(2)  # 페이지 로드 대기

    # "더보기" 버튼 클릭
    more_button_xpath = f"/html/body/ytd-app/div[1]/ytd-page-manager/ytd-shorts/div[3]/div/ytd-reel-video-renderer[{i+1}]/div[4]/ytd-reel-player-overlay-renderer/div[2]/div/div[4]"
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, more_button_xpath))).click()
    time.sleep(5)
    # "추가설명창" 열기
    WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="items"]/ytd-menu-service-item-renderer[1]/tp-yt-paper-item'))
    ).click()
    print('추가설명창 열림')
    time.sleep(5)

    
    # 태그, 조회수, 업로드 날짜 수집
    tags_elements = driver.find_elements(By.XPATH, '//*[@id="description"]/yt-formatted-string/a')
    tags_text = [tag.text.replace('#', '') for tag in tags_elements]
    views_text = driver.find_element(By.XPATH, '//*[@id="factoids"]/factoid-renderer[2]/div/span[1]/span').text
    upload_date_text = driver.find_element(By.XPATH, '//*[@id="factoids"]/factoid-renderer[3]/div/span[1]/span').text

    # 데이터 리스트에 추가
    all_video_data.append([titles[i], tags_text, views_text, upload_date_text])
    print('*' * 100)
    print(all_video_data[-1])
    print('*' * 100)
    # 뒤로 가기
    driver.back()
    time.sleep(3)  # 페이지 로드 대기
    
    # 데이터 리스트에 추가
    all_video_data.append([i + 1, titles[i], tags_text, views_text, upload_date_text])
    print('*' * 100)
    print(all_video_data[-1])
    print('*' * 100)
    # 뒤로 가기
    driver.back()
    time.sleep(3)  # 페이지 로드 대기

# DataFrame 생성
df = pd.DataFrame(all_video_data, columns=['순서', '제목', '태그', '조회수', '업로드 날짜'])

# Excel과 CSV 파일로 저장
df.to_excel('youtube_data.xlsx', index=False)
df.to_csv('youtube_data.csv', index=False)

print('-' * 100)
print("Data saved to 'Gametalktalk_data.xlsx' and 'Gametalktalk_data.csv'")
print('-' * 100)
```
