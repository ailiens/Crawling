```python
# CSS Selector로 교체하기
###### https://www.youtube.com/@hare_u/shorts

from selenium.common import TimeoutException
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Selenium WebDriver 설정 (Chrome 사용)
s = Service("C:\\chromedriver\\chromedriver.exe")
driver = webdriver.Chrome(service=s)

# YouTube에 접속
driver.get("https://www.youtube.com/@GameTalkTalk/shorts")
time.sleep(2)  # 페이지 로드를 위한 대기 시간

# Shorts 탭 클릭
shorts_button = driver.find_element(By.XPATH, '//*[@id="tabsContent"]/yt-tab-group-shape/div[1]/yt-tab-shape[3]/div[1]')
shorts_button.click()
time.sleep(2)

titles_elements = driver.find_elements(By.XPATH, '//*[@id="video-title"]')
titles = [title.text.strip() if title.text.strip() else "직접기입" for title in titles_elements]
# 데이터 저장을 위한 리스트 초기화
all_video_data = []

# 각 비디오 별로 추가 정보 수집
videos_xpath = '//*[@id="dismissible"]'
videos = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, videos_xpath)))
print('제목 개수: ',len(titles), '비디오 개수: ',len(videos))

for i, video in enumerate(videos):
    video.click()  # 비디오 클릭
    time.sleep(2) # 페이지 로드 대기

# for i, video in range(45):
#     video[i].click()  # 비디오 클릭
#     time.sleep(2) # 페이지 로드 대기    
    
    # "더보기" 버튼 클릭
    more_button_xpath = f"/html/body/ytd-app/div[1]/ytd-page-manager/ytd-shorts/div[3]/div/ytd-reel-video-renderer[{i+1}]/div[4]/ytd-reel-player-overlay-renderer/div[2]/div/div[4]"
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, more_button_xpath))).click()
    time.sleep(5)
    print('worked')
    # "추가설명창" 열기
    WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, '#items > ytd-menu-service-item-renderer.style-scope.ytd-menu-popup-renderer.iron-selected'))
    ).click()
    print('추가설명창 열림')
    time.sleep(5)
# 
#     # 태그, 조회수, 업로드 날짜 수집
    tags_elements = driver.find_elements(By.CSS_SELECTOR, '#description > yt-formatted-string > a')
    tags_text = [tag.text.replace('#', '') for tag in tags_elements]
    like = driver.find_element(By.CSS_SELECTOR, '#factoids > factoid-renderer:nth-child(1) > div > span.YtwFactoidRendererValue > span').text
    views_text = driver.find_element(By.CSS_SELECTOR, '#factoids > factoid-renderer:nth-child(2) > div > span.YtwFactoidRendererValue > span').text
    upload_date_text = driver.find_element(By.CSS_SELECTOR, '#factoids > factoid-renderer:nth-child(3) > div').text
    upload_date = upload_date_text.replace('\n', ' ')
    
# 
    # 데이터 리스트에 추가
    all_video_data.append([i + 1, titles[i], tags_text, like, views_text, upload_date])
    print('*' * 100)
    print(all_video_data[-1])
    print('*' * 100)
    # 뒤로 가기
    driver.back()
    time.sleep(3) # 페이지 로드 대기
    # print('뒤로가가기')
# 
# DataFrame 생성
df = pd.DataFrame(all_video_data, columns=['순서', '제목', '태그','좋아요', '조회수', '업로드 날짜'])

# Excel과 CSV 파일로 저장
df.to_excel('0507_Gamtalktalk_youtube_data.xlsx', index=False)
df.to_csv('0507_Gamtalktalk_youtube_data.csv', index=False)

print('-' * 100)
print("Data saved to '겜톡톡' 데이터 크롤링 완료")
print('-' * 100)
```

```python
file_path = '0507_Gamtalktalk_youtube_data.xlsx'
data = pd.read_excel(file_path)

def convert_korean_number_to_integer(korean_number):
    try:
        # '천'이 포함된 경우, 예를 들어 '1.2천' -> 1200으로 변환
        if '천' in korean_number:
            number = float(korean_number.replace('천', ''))
            return int(number * 1000)
        else:
            # 숫자로 직접 변환을 시도합니다.
            return int(korean_number)
    except ValueError:
        # 숫자로 변환할 수 없는 경우 0 또는 기본값으로 처리
        return 0


# 날짜 변환 함수 정의
def parse_date(date_str):
    try:
        return pd.to_datetime(date_str, format='%m월 %d일 %Y년', errors='raise')
    except ValueError:
        return pd.to_datetime(date_str, format='%Y년 %m월 %d일', errors='raise')

# '업로드 날짜' 컬럼의 날짜 형식을 확인하고 datetime 객체로 변환
data['업로드 날짜'] = data['업로드 날짜'].apply(parse_date)

# 요일 정보를 한글로 설정 (datetime 객체 사용 전에 수행)
days = ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일']
data['요일'] = data['업로드 날짜'].dt.dayofweek.map(lambda x: days[x])

# 시간 정보를 제거하고 날짜만 남김
data['업로드 날짜'] = data['업로드 날짜'].dt.date

# 조회수 데이터를 숫자 형식으로 변환하기 전에 한국어 수치를 정수로 변환
# data['조회수'] = data['조회수'].apply(convert_korean_number_to_integer)
data['좋아요'] = data['좋아요'].apply(convert_korean_number_to_integer)
                                


# 데이터의 첫 몇 줄을 확인
print(data.head())
```

2. 상호 작용의 이점: '좋아요'와 '조회수' 사이의 상관관계는 두 메트릭 간의 상호 작용을 시사합니다. 비디오가 좋아요를 많이 받으면 알고리즘이 이를 더 많은 사람들에게 추천할 가능성이 높아지고, 결과적으로 조회수가 증가합니다. 반대로, 많은 조회수를 얻은 비디오는 더 많은 좋아요를 얻을 기회가 많아집니다.
3. 인기도와 참여도: 조회수는 비디오가 얼마나 많이 시청되었는지를 나타내며, 좋아요 수는 시청자들의 참여도와 만족도를 반영합니다. 두 변수 모두 비디오의 인기도를 나타내는 중요한 지표로, 높은 상관관계는 비디오 콘텐츠가 시청자에게 긍정적인 반응을 이끌어내고 있다는 강력한 신호를 보냅니다.

```python
# Preparing data for tag analysis with correct parsing
from collections import defaultdict
import ast

# Initializing dictionaries for counting and summing likes/views per tag
tag_counts = defaultdict(int)
tag_likes = defaultdict(int)
tag_views = defaultdict(int)

# Process each row in the data, correctly parsing the tags
for _, row in filtered_data.iterrows():
    try:
        tags = ast.literal_eval(row['태그'])  # Using ast.literal_eval to convert string to list
        for tag in tags:
            tag_counts[tag] += 1
            tag_likes[tag] += row['좋아요']
            tag_views[tag] += row['조회수']
    except:
        continue  # Skip rows where the '태그' column cannot be parsed

# Calculating average likes and views per tag
tag_performance = pd.DataFrame({
    "Tag": list(tag_counts.keys()),
    "Count": list(tag_counts.values()),
    "Average Likes": [tag_likes[tag] / tag_counts[tag] for tag in tag_counts],
    "Average Views": [tag_views[tag] / tag_counts[tag] for tag in tag_counts]
}).sort_values(by="Count", ascending=False).reset_index(drop=True)

# Convert average likes and views to integers
tag_performance['Average Likes'] = tag_performance['Average Likes'].astype(int)
tag_performance['Average Views'] = tag_performance['Average Views'].astype(int)

# Showing the top 10 tags by count
top_tags = tag_performance.head(10)
print(top_tags)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 조회수 컬럼의 한글 수치 변환
def convert_korean_number_to_integer(korean_number):
    if isinstance(korean_number, str):
        korean_number = korean_number.replace(',', '')  # 콤마 제거
        if '천' in korean_number:
            number = float(korean_number.replace('천', ''))
            return int(number * 1000)
        return int(korean_number)
    return korean_number

# 데이터 로드 및 전처리
data['업로드 날짜'] = pd.to_datetime(data['업로드 날짜'], errors='coerce')  # 날짜 변환
# 조회수 데이터 변환
data['조회수'] = data['조회수'].apply(convert_korean_number_to_integer)
# 조회수가 높은 상위 10개의 영상 데이터 추출
top_videos_by_views = data.nlargest(10, '조회수')

# 수평 바 차트로 시각화
plt.figure(figsize=(12, 8))
plt.barh(top_videos_by_views['제목'], top_videos_by_views['조회수'], color='blue')
plt.ylabel('Video Titles', fontsize=14)
plt.xlabel('Average Views', fontsize=14)
plt.title('Top 10 Videos by Views', fontsize=16)
plt.gca().invert_yaxis()  # 타이틀을 높은 조회수부터 위로 표시
plt.tight_layout()
plt.show()
```
